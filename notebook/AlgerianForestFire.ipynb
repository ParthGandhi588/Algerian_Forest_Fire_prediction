```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Header = 1 will ignore the first row as header
df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Complete-Data-Science-With-Machine-Learning-And-NLP-2024/refs/heads/main/3-Complete%20Linear%20Regression/Practicals/Algerian_forest_fires_dataset_UPDATE.csv', header = 1)
df.head()
```

```python
df.info()
```

### Data Cleaning

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122, "Region"] = 0
df.loc[122:, "Region"] = 1
```

```python
df[['Region']] = df[['Region']].astype(int)
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.loc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df[['month','day','year','Temperature','RH','Ws']] = df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
objects = [features for features in df.columns if df[features].dtype == 'O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype('float')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv('AlgerianForestFire_cleaned_Dataset.csv', index=False)
```

### EDA

```python
df_copy = df.drop(['day','month','year'], axis=1)
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Encodeing of the category classes
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Plot density for all features
plt.style.use('ggplot')
df_copy.hist(bins = 50, figsize=(20,15))
plt.show()
```

```python
## Percentage of pie chart

percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
classlabels = ['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title("Pie chart of Classes")
plt.show()
```

```python
## Correlation
corr = df_copy.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(corr, annot=True)
```

```python
## BOX plot for outlier
sns.boxplot(df_copy['FWI'])
```
