```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('AlgerianForestFire_cleaned_Dataset.csv')
df.head()
```

```python
df.columns
```

```python
df.drop(columns=['day','month','year'], axis=1, inplace=True)
df.head()
```

```python
df['Classes'] = np.where(df['Classes'].str.contains("not fire"), 0, 1)
```

```python
df['Classes'].value_counts()
```

```python
## Independent and Dependent Features
X = df.drop('FWI', axis=1)
y = df['FWI']
```

```python
## Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
## Feature Selection
X_train.corr()
```

```python
## Check for Multicollinearity
plt.figure(figsize=(15,10))
corr = X_train.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j] > threshold):
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train,0.85)
```

```python
X_train.drop(corr_features, axis=1, inplace=True)
X_test.drop(corr_features, axis=1, inplace=True)
X_train.shape, X_test.shape
```

```python
## Feature Scalling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
## Boxplots to understand effect of Standard Scaler
plt.subplots(figsize=(15,10))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title("Before Standard Scaler")
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title("After Standard Scaler")
```

##  Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
Regression = LinearRegression()
```

```python
Regression.fit(X_train_scaled, y_train)
y_pred = Regression.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
r2_score = r2_score(y_test, y_pred)
print(mae)
print(r2_score)
plt.scatter(y_test,y_pred)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
Lasso = Lasso()
Lasso.fit(X_train_scaled, y_train)
y_pred = Lasso.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
r2_score = r2_score(y_test, y_pred)
print(mae)
print(r2_score)
plt.scatter(y_test,y_pred)
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
Ridge = Ridge()
Ridge.fit(X_train_scaled, y_train)
y_pred = Ridge.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
r2_score = r2_score(y_test, y_pred)
print(mae)
print(r2_score)
plt.scatter(y_test,y_pred)
```

## Elastic Net Regression

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ElasticNet = ElasticNet()
ElasticNet.fit(X_train_scaled, y_train)
y_pred = ElasticNet.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
r2_score = r2_score(y_test, y_pred)
print(mae)
print(r2_score)
plt.scatter(y_test,y_pred)
```

## Cross Validation LassoCV()

```python
from sklearn.linear_model import LassoCV
lassocv = LassoCV(cv = 5)
lassocv.fit(X_train_scaled, y_train)
```

```python
y_pred = lassocv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
```

## RidgeCV

```python
from sklearn.linear_model import RidgeCV
ridgecv = RidgeCV(cv=5)
ridgecv.fit(X_train_scaled, y_train)
y_pred = ridgecv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import ElasticNetCV
elasticnetcv = ElasticNetCV()
elasticnetcv.fit(X_train_scaled, y_train)
y_pred = elasticnetcv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
```
